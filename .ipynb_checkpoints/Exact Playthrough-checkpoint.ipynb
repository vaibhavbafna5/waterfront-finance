```python
import locale
locale.setlocale(locale.LC_ALL, '')  # Use '' for auto, or force e.g. to 'en_US.UTF-8'

# @CHRISTIAN
# - change the interest rate to understand more punishing scenarios
# - change the revenue & profit percentages to reflect different businesses
# - change the OWED variable to choose a different loan amount

# 6.8% interest
INTEREST = 1.068

# revenue
REVENUE = 364000

# profit as a percentage of revenue
PROFIT = 0.15 * REVENUE # % margins on revenue

# loan amount
OWED = 420000

print("interest:", f'{(INTEREST - 1) * 100}%')
print("yearly revenue:", '${:n}'.format(REVENUE))
print("yearly profit:", '${:n}'.format(PROFIT))

for x in range(1, 25):
    
    OWED = OWED - PROFIT
    
    if OWED <= 0:
        print("loan free in year: ", x)
        break
        
    OWED = OWED * INTEREST
    
    print("end of year: ", x)
    print("amt owed: ", OWED)
    print("\n")
```
