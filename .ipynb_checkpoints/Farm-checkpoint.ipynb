```python
interest = 1.068 # 6% interest
revenue = 364000 # 720,000 revenue per year
profit = 0.15 * revenue # % margins on revenue

owed = 420000 # mortage amount

print("interest:", interest - 1)
print("yearly revenue:", '{:n}'.format(revenue))
print("yearly profit:", '{:n}'.format(profit))
```

```python
for x in range(1, 25):
    
    owed = owed - profit
    
    if owed <= 0:
        print("loan free in year: ", x)
        break
        
    owed = owed * interest
    
    print("end of year: ", x)
    print("amt owed: ", owed)
    print("\n")
```

```python
possible_interest_rates = [(1 + (x / 100)) for x in range(6, 10)]
for i in possible_interest_rates:
    print("possible interest rate (%):", int((i - 1) * 100))
```

```python
revenues = [x * 30000 for x in range(1, 20)]
for revenue in revenues:
    print('${:n}'.format(revenue))
```

```python
combos = []

for interest in possible_interest_rates:
    for revenue in revenues:

        interest = interest # 6% interest
        revenue = revenue # 720,000 revenue per year
        profit_percentage = 0.18 # 10% profit
        owed = 420000 # mortage amount
        threshold = 15

        for x in range(1, 50):

            profit = profit_percentage * revenue # % margins on revenue
            owed = owed - profit

            if owed <= 0 and x <= threshold:
                combos.append((interest, revenue, x))
                print("loan free in year: ", x)
                break

            owed = owed * interest
```

```python
print("possible combinations that let you even out in year:", threshold, "\n")

for tup in combos:
    print("interest:", (tup[0] - 1) * 100)
    print("revenue:", '{:n}'.format(tup[1]))
    print("average yearly profit:", '${:n}'.format(tup[1] * profit_percentage))
    print("break even year:", tup[2])
    print("\n")
```

```python
import random

N = 1000000

revenues = []
years = []


for i in range(0, N):

    random_revenue = random.randint(10000, 500000)


    interest = 1.068 # 6% interest
    revenue = random_revenue # 720,000 revenue per year
    profit_percentage = 0.18 # 10% profit
    owed = 420000 # mortage amount
    threshold = 300

    for x in range(1, 300):

        profit = profit_percentage * revenue # % margins on revenue
        owed = owed - profit

        if owed <= 0 and x <= threshold:
            revenues.append(profit_percentage * revenue)
            years.append(x)
    #         print("loan free in year: ", x)
            break

        owed = owed * interest
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(revenues, years)
plt.show()
```
