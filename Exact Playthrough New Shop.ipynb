```python
import locale
locale.setlocale(locale.LC_ALL, '')  # Use '' for auto, or force e.g. to 'en_US.UTF-8'

# @CHRISTIAN
# - The question here is when you will be net positive after initial expenses & monthly costs
# - only change variables in the outlined region
# --------------------------------------------------
# --------------------------------------------------

# fixed amount you pay to get bike rentals
RENTALS = 50000

# fixed amount you pay to set up repair equipment
REPAIRS = 100000

# monthly rent
RENT = 6000 

# monthly insurance
INSURANCE = 1000

# montly payment for utilities (internet, software, etc)
UTILITIES = 1000

# profit per month
PROFIT_PER_MONTH = 10000

# --------------------------------------------------
# --------------------------------------------------

initial_expenses = RENTALS + REPAIRS
monthly_cost = RENT + INSURANCE + UTILITIES

print("INITIAL EXPENSES:", '${:n}'.format(initial_expenses))
print("MONTHLY EXPENSES:", '${:n}'.format(monthly_cost))
print("\n")

down = initial_expenses
for i in range(0, 120):
    down = (down + monthly_cost) - PROFIT_PER_MONTH
    if down < 0:
        print(f"\nturn a profit in month {i} of operation")
        break
    print(f"money you owe in month {i} of operation: ", '${:n}'.format(down))
```
